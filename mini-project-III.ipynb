```python
import pandas as pd
import plotly.express as px
from sklearn.decomposition import PCA
import numpy as np
```

```python
# inputs 
# DF dataframe to edit, 
# STR heading in df where date is stored, 
# STR new heading for month name, 
# STR new heading for day name,
# STR new heading for year name
# output
# new DF with new columns for date and remove existing date column 
def date_split (df,heading,headmonth,headday,headyear):
    x = list(df[heading])
    month = []
    day = []
    year = []
    for i in range(len(x)):
        month.append(x[i].split('.')[0])
        day.append(x[i].split('.')[1])
        year.append(x[i].split('.')[2])
    df[headmonth] = month
    df[headday] = day            
    df[headyear] = year
    df.drop(heading,axis = 1, inplace = True)
    return (df)
```
